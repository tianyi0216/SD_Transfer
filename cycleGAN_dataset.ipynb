## Code for CycleGAN Data Generation

TO run this file, first clone the repository from cycleGAN by

```git clone https://github.com/junyanz/pytorch-CycleGAN-and-pix2pix```

Note: please also install the required packages as instructed in the repository.

Then, naviagte to the directory, put this file in and just run the notebook.

````python
# import statements
import os
from subprocess import check_output
````

````python
avail_models = "summer2winter_yosemite, winter2summer_yosemite, style_monet, style_cezanne, style_ukiyoe, style_vangogh".split(", ")
````

````python
avail_datasets = "summer2winter_yosemite, monet2photo, cezanne2photo, ukiyoe2photo, vangogh2photo".split(", ")
````

````python
# example download: bash ./datasets/download_cyclegan_dataset.sh horse2zebra

for dataset in avail_datasets:
    out = check_output(["bash", "./datasets/download_cyclegan_dataset.sh", dataset])

print(out)
````

````python
# download pretrained models
for model in avail_models:
    out = check_output(["bash", "./scripts/download_cyclegan_model.sh", model])

print(out)
````

````python
# run each model

#example: python test.py --dataroot datasets/horse2zebra/testA --name horse2zebra_pretrained --model test --no_dropout
out = check_output(["python", "test.py", "--dataroot", "datasets/summer2winter_yosemite/trainA", "--name", "summer2winter_yosemite_pretrained", "--model", "test", "--no_dropout", "--num_test", "500", "--gpu_ids", "-1"])
out = check_output(["python", "test.py", "--dataroot", "datasets/summer2winter_yosemite/trainB", "--name", "winter2summer_yosemite_pretrained", "--model", "test", "--no_dropout", "--num_test", "500", "--gpu_ids", "-1"])
out = check_output(["python", "test.py", "--dataroot", "datasets/monet2photo/trainB", "--name", "style_monet_pretrained", "--model", "test", "--no_dropout", "--num_test", "500", "--gpu_ids", "-1"])
out = check_output(["python", "test.py", "--dataroot", "datasets/cezanne2photo/trainB", "--name", "style_cezanne_pretrained", "--model", "test", "--no_dropout", "--num_test", "1000", "--gpu_ids", "-1"])
out = check_output(["python", "test.py", "--dataroot", "datasets/ukiyoe2photo/trainB", "--name", "style_ukiyoe_pretrained", "--model", "test", "--no_dropout", "--num_test", "1500", "--gpu_ids", "-1"])
out = check_output(["python", "test.py", "--dataroot", "datasets/vangogh2photo/trainB", "--name", "style_vangogh_pretrained", "--model", "test", "--no_dropout", "--num_test", "2000", "--gpu_ids", "-1"])
````

````python
# move results to a new directory called custom_datasets

if not os.path.exists("custom_datasets"):
    os.makedirs("custom_datasets")
if not os.path.exists("custom_datasets/source"):
    os.makedirs("custom_datasets/source")
if not os.path.exists("custom_datasets/target"):
    os.makedirs("custom_datasets/target")
````

````python
# copy the 500 images from winter2summer_yosemite to source and target, rename them to 1.png, 2.png, 3.png, etc.

all_images = os.listdir("results/winter2summer_yosemite_pretrained/test_latest/images")

real_images = [x for x in all_images if "real" in x]
fake_images = [x for x in all_images if "fake" in x]

# sort the images
real_images.sort()
fake_images.sort()

for i in range(1, 501):
    # check if real and fake images are the same
    if real_images[i-1].split("_")[0] != fake_images[i-1].split("_")[0]:
        print("Error: real and fake images do not match")
        print("real image is: " + real_images[i-1] + " and fake image is: " + fake_images[i-1])
        continue
    
    # copy the images
    os.rename("results/winter2summer_yosemite_pretrained/test_latest/images/" + real_images[i-1], "custom_datasets/source/" + str(i) + ".png")
    os.rename("results/winter2summer_yosemite_pretrained/test_latest/images/" + fake_images[i-1], "custom_datasets/target/" + str(i) + ".png")
````
