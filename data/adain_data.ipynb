````python
# import statements
import os
from subprocess import check_output
import random
````

````python
avail_style = ["monet", "vangogh", "cezanne", "ukiyoe", "degas", "picasso"]
````

````python
# I want 3000 training images and 280 validation images
training_dir_1 = "00000"
training_dir_2 = "00003"
training_dir_3 = "00010"
training_dir_4 = "00011"
training_dir_5 = "00018"
validation_dir = "00001"

target_training_dir = "training_set"
target_validation_dir = "testing_set"
````

````python
# find all jpg files in the directory
training_jpg_files_1 = sorted([f for f in os.listdir(training_dir_1) if f.endswith(".jpg")])
training_jpg_files_2 = sorted([f for f in os.listdir(training_dir_2) if f.endswith(".jpg")])
training_jpg_files_3 = sorted([f for f in os.listdir(training_dir_3) if f.endswith(".jpg")])
training_jpg_files_4 = sorted([f for f in os.listdir(training_dir_4) if f.endswith(".jpg")])
training_jpg_files_5 = sorted([f for f in os.listdir(training_dir_5) if f.endswith(".jpg")])
validation_jpg_files = sorted([f for f in os.listdir(validation_dir) if f.endswith(".jpg")])
````

````python
# only consider images divisible by 10
val_path = []
for i in range(0, len(validation_jpg_files), 10):
    val_path.append(os.path.join(validation_dir, validation_jpg_files[i]))

train_1_path = []
for i in range(0, len(training_jpg_files_1), 10):
    train_1_path.append(os.path.join(training_dir_1, training_jpg_files_1[i]))

train_2_path = []
for i in range(0, len(training_jpg_files_2), 10):
    train_2_path.append(os.path.join(training_dir_2, training_jpg_files_2[i]))

train_3_path = []
for i in range(0, len(training_jpg_files_3), 10):
    train_3_path.append(os.path.join(training_dir_3, training_jpg_files_3[i]))

train_4_path = []
for i in range(0, len(training_jpg_files_4), 10):
    train_4_path.append(os.path.join(training_dir_4, training_jpg_files_4[i]))

train_5_path = []
for i in range(0, len(training_jpg_files_5), 10):
    train_5_path.append(os.path.join(training_dir_5, training_jpg_files_5[i]))
````

````python
# move the training images to the training set
import shutil

for i in range(len(train_1_path)):
    shutil.copy(train_1_path[i], os.path.join(target_training_dir, train_1_path[i].split("/")[1]))
    shutil.copy(train_1_txt[i], os.path.join(target_training_dir, train_1_txt[i].split("/")[1]))

for i in range(len(train_2_path)):
    shutil.copy(train_2_path[i], os.path.join(target_training_dir, train_2_path[i].split("/")[1]))
    shutil.copy(train_2_txt[i], os.path.join(target_training_dir, train_2_txt[i].split("/")[1]))

for i in range(len(train_3_path)):
    shutil.copy(train_3_path[i], os.path.join(target_training_dir, train_3_path[i].split("/")[1]))
    shutil.copy(train_3_txt[i], os.path.join(target_training_dir, train_3_txt[i].split("/")[1]))

for i in range(len(train_4_path)):
    shutil.copy(train_4_path[i], os.path.join(target_training_dir, train_4_path[i].split("/")[1]))
    shutil.copy(train_4_txt[i], os.path.join(target_training_dir, train_4_txt[i].split("/")[1]))

for i in range(len(train_5_path)):
    shutil.copy(train_5_path[i], os.path.join(target_training_dir, train_5_path[i].split("/")[1]))
    shutil.copy(train_5_txt[i], os.path.join(target_training_dir, train_5_txt[i].split("/")[1]))
````
