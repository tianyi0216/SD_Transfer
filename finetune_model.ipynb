```python
%%bash
export MODEL_ID="runwayml/stable-diffusion-v1-5"
export DATASET_ID="HL121/stat453_dataset"
export OUTPUT_DIR="453model_1"

accelerate launch --mixed_precision="fp16" train_instruct_pix2pix.py \
  --pretrained_model_name_or_path=$MODEL_ID \
  --dataset_name=$DATASET_ID \
  --use_ema \
  --resolution=256 --random_flip \
  --train_batch_size=2 --gradient_accumulation_steps=4 --gradient_checkpointing \
  --max_train_steps=750 \
  --checkpointing_steps=5000 --checkpoints_total_limit=1 \
  --learning_rate=5e-05 --lr_warmup_steps=0 --conditioning_dropout_prob=5e-2 \
  --mixed_precision=fp16 \
  --val_image_url="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTFKrbTgG508DGYc3_RyZAsqnDazwpwmZAo64gdE7tO3A&s" \
  --validation_prompt="Transfer to vangogh style" \
  --seed=42 \
  --output_dir=$OUTPUT_DIR \
  --report_to=wandb \
```
